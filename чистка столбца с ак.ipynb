```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('добавил столбец аминокислот (нужно почистить).xlsx')
```

```python
def a(line):
    print(line,'\n', type(line))
    if isinstance(line, str) and line[:3]=="ID:":
        try:
            print('победа')
            return(line[ line.find('Seq(')+5:-2] )
        except:
            print('неудача')
            return line
```

```python
for i in range(10):
    print(df['АК последовательность белка-мишени'][i])
```

```python
df['последовательности почищенные'] = df['АК последовательность белка-мишени'].apply(a)
```
