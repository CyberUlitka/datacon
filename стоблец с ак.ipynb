```python
import pandas as pd
import numpy as np

from rdkit import Chem

from Bio import Entrez
from Bio import SeqIO
```

```python
df = pd.read_excel('UTexas Aptamer Database.xlsx')
```

```python
df.columns = df.columns.str.strip()
```

```python
def to_numeric(column):
    return pd.to_numeric(df[column], errors='coerce')

df['Kd (nM)'] = to_numeric('Kd (nM)')
df['Sequence Length'] = to_numeric('Sequence Length')
df['GC Content'] = to_numeric('GC Content')
```

```python
Entrez.email = "arsendoronin@gmail.com"
```

```python
last=''
lastseq=''

def fetch_protein_sequence(protein_name):
    global last
    global lastseq

    if last==protein_name:
        print('повторка')
        return lastseq
    last=protein_name

    handle = Entrez.esearch(db="protein", term=protein_name, retmax=1)
    record = Entrez.read(handle)

    if record['IdList']:
        protein_id = record['IdList'][0]
        handle = Entrez.efetch(db="protein", id=protein_id, rettype="fasta", retmode="text")
        seq_record = SeqIO.read(handle, "fasta")
        print(protein_name, ' найден, ура-ура')
        
        lastseq=seq_record
        
        handle.close()
        return seq_record.seq
    else:
        print(protein_name, ' не найден')
        lastseq=None
        return None

df['АК последовательность белка-мишени'] = df['Target'].apply(fetch_protein_sequence)
```

```python
df.to_excel('output.xlsx', index=False)
```
